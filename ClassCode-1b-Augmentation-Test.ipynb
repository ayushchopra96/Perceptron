```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im = cv2.imread('/home/ayush/Desktop/cat.jpg')
print im.shape
plt.imshow(im)
plt.show()
```

```python
im = cv2.resize(im,(150,150))
print im.shape
im = im.reshape((1,)+im.shape)
print im.shape
```

```python
datagen = ImageDataGenerator(
    #horizontal_flip=True,
    height_shift_range=0.6)
    #rotation_range=60)
```

```python
import os
os.mkdir('./flow_test_class13')
```

```python
i=0
for batch in datagen.flow(im,batch_size=2,save_to_dir='./flow_test_class13/',save_format='jpeg',
                          save_prefix='class12'):
    i+=1
    if i>2:
        break
```

```python
tested_imgs = os.listdir('./flow_test_class12/')
print len(tested_imgs)
```

```python
base_path = './flow_test_class12/'
for ix in tested_imgs:
    path = base_path+ix
    ig = cv2.imread(path)
    plt.imshow(ig)
    plt.show()
```

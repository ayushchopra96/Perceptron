```python
#os.mkdir('./trainset/')
def getname(ind):
    if ind==1:
        return 'dog'
    else:
        return 'cat'
```

```python
dog_count = 0
cat_count = 0
x_train_cat = []
x_train_dog = []
y_train_cat = []
y_train_dog = []
y_train = []

max_cat = 1000
max_dog = 1000
for ix in range(0,len(imgs)):
    if dog_count < max_dog and cat_count< max_cat:
        ig = imgs[ix]
        ind = index(ig.split('.')[0])
        y_train.append(ind)
        path = './train/' + ig
        im = cv2.imread(path)
        im = cv2.resize(im,(150,150))
        name = getname(ind)
        if name=='dog':
            new_path = './trainset_dog/' + 'dog_'+str(dog_count)+'.jpeg'
        else:
            new_path = './trainset_cat/'+'cat_' + str(cat_count)+'.jpeg'
        cv2.imwrite(new_path,im)
        print im.shape
        if ind==1:
            x_train_dog.append(im)
            dog_count+=1
        else:
            x_train_cat.append(im)
            cat_count+=1
        #print im.shape
    elif dog_count<max_dog:
        ig = imgs[ix]
        ind = index(ig.split('.')[0])
        if ind==1:
            y_train.append(ind)
            path = './train/' + ig
            im = cv2.imread(path)
            im = cv2.resize(im,(150,150))
            new_path = './trainset_dog/'+ 'dog_'+ str(dog_count)+'.jpeg'
            cv2.imwrite(new_path,im)
            print im.shape
            x_train_dog.append(im)
            dog_count+=1
        else:
            continue
            
    elif cat_count<max_cat:
        ig = imgs[ix]
        ind = index(ig.split('.')[0])
        if ind==0:
            y_train.append(ind)
            path = './train/' + ig
            im = cv2.imread(path)
            im = cv2.resize(im,(150,150))
            new_path = './trainset_cat/'+'cat_'+str(cat_count)+'.jpeg'
            cv2.imwrite(new_path,im)
            print im.shape
            x_train_cat.append(im)
            cat_count+=1
        else:
            continue
```

Normalize the Data<br>
Split into test and train sets<br>
make npy files

```python
#Normalisation
Xcat = np.asarray(x_train_cat)
Xdog = np.asarray(x_train_dog)

Xcat = Xcat/255.0
Xdog = Xdog/255.0 

print Xcat.shape, Xdog.shape
```

```python
#split into test and train set
split = int(0.7*Xcat.shape[0])

XtrainCat = Xcat[:split]
XtestCat = Xcat[split:]

XtrainDog = Xdog[:split]
XtestDog = Xdog[split:]

print XtrainCat.shape, XtrainDog.shape
```

```python
x_train = np.vstack((XtrainCat,XtrainDog))
y_train = np.zeros(XtrainCat.shape[0]+XtrainDog.shape[0])

y_train[XtrainCat.shape[0]:XtrainDog.shape[0]+XtrainCat.shape[0]]=1
print x_train.shape
print y_train.shape


x_test = np.vstack((XtestCat,XtestDog))
y_test = np.zeros(XtestCat.shape[0]+XtestDog.shape[0])

y_test[XtestDog.shape[0]:XtestCat.shape[0]+XtestDog.shape[0]]=1

print x_test.shape
print y_test.shape
```
